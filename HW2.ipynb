## Задание 1
Дан текст. Необходимо выделить все биграммы в этом тексте и сохранить их список в какую-либо переменную. **Биграммы** -- это пары стоящих рядом слов, например, "варкалось хливкие", "хливкие шорьки" и т.д.

**Подсказка №1**: текст нужно разбить на слова с помощью функции split(), а потом использовать цикл for для создания списка биграмм. На каждой итерации цикла нужно печатать биграмму, а потом добавлять ее в список.

**Подсказка №2**: количество биграмм в тексте = количество слов в тексте - 1.

```python
a = """его кошка проснулась на столе его кошка боится"""
words = a.split()
bigrams = []
for i in range (len (words)-1):
    bigrams.append([words[i],words[i+1]])
print(bigrams)
```
