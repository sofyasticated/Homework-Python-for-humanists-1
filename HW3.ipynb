## Задание 1
Скачайте текст ["Литературных анекдотов"](https://github.com/ancatmara/python-for-dh/blob/master/Classes/9-10/literary_anecdotes.txt). Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
#делаем функцию 
def function (text,new_text):
    with open(text,'r',encoding='utf-8') as g:
        read_text = g.read() #фунция читает аргумент 
        #функция убирает пунктуацию для дальнейшей лемматизации 
        words = [word.strip("""!?.«»():"";,—-""") for word in read_text.split()]
        words = [word for word in words if len(word)>0]
        text_bez_punkt =' '.join(words)
        #лемматизация 
        from pymystem3 import Mystem 
        m = Mystem()
        lemmas_0= m.lemmatize(text_bez_punkt)
        lemmas= ''.join(lemmas_0)
        #записываем в новый файл
        with open(new_text,'w',encoding='utf-8') as h:
            for l in lemmas:
                h.write (lemmas)
    return lemmas
lemmas = function ('literary_memes.txt', 'memes_7.txt')
```

## Задание 2 
Очистите лемматизированный текст от стоп-слов и посчитайте ipm для оставшихся. Выведите 20 самых частотных по этому параметру слов.

```python
#читаем стоп-слова
import os
os.chdir (r'D:\literature')
with open('rus_stopwords.txt','r',encoding='utf-8') as n:
        m = n.read()
        stop_slova = m.split()
        text_stop = []
        spisok_lem = lemmas.split()
        for i in spisok_lem:
            if i not in stop_slova:
                text_stop.append(i)
                str_text_stop=' '.join(text_stop)
#считаем общее количество слов
from collections import Counter
counts = Counter(str_text_stop.split())
mnogo = len(spisok_lem)
spisok = []
for s in counts.items():
    spisok.append((s[0], s[1] / mnogo* 1000000)) #считаем ipm
sorted_spisok = sorted(spisok, key=lambda x: x[1], reverse=True) #сортируем, чтобы вывести 20 самых частотных слов
print(sorted_spisok[:20])
```

## Задание 3
Сделайте полный морфологический разбор исходного текста. Напишите регулярное выражение, которое будет извлекать из тега только часть речи. Пройдитесь циклом по списку с разборами, который выдал pymystem3, извлекая из каждого разбора форму слова и его часть речи и записывая их в новый словарь (форма -- ключ, часть речи -- значение). Посчитайте абсолютную частоту для всех частей речи, а затем относительнную частоту (абсолютная частота / длина текста в словах).

```python
text_bez_punkt = function ('literary_memes.txt', 'memes_7.txt') #чтобы получить исходный текст 
from pymystem3 import Mystem 
m = Mystem()
kuzdra = m.analyze(text_bez_punkt)# морфологический разбор текста 
parts = []
import re
def get(s): #функция с регулярным выражением
    pattern= re.match('[A-Z]+',s).group(0)
    return (pattern) #регулярное выражение - из тега только часть речи извлекает 
for word in kuzdra:
    try:
        parts.append(get(word['analysis'][0]['gr']))
    except (KeyError, IndexError) as e:  
        pass
print (parts)
```

```python
dictionary = {}
for word in kuzdra:
    try:
        dictionary[word['text']]= get(word['analysis'][0]['gr'])
    except (KeyError, IndexError) as e:  
        pass
print(dictionary)
```

```python
absolut_counts = Counter(parts)
print(absolut_counts)
mnogo = len(parts)
for s in absolut_counts.items():
    print(s[0], s[1] / mnogo)
```
